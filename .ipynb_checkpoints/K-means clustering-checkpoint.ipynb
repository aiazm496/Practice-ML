```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#generate artificial data
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features= 2 , centers= 4 , cluster_std= 1.8, random_state=101)
```

```python
feature_1 = data[0][:,0]
feature_1
```

```python
feature_2 = data[0][:,1]
feature_2
```

```python
sns.set_style('darkgrid')
sns.scatterplot(feature_1,feature_2, hue=data[1], cmap = 'rainbow')
```

```python
# 4 blobs visible
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=4)
```

```python
km.fit(data[0])
```

```python
km.labels_  #predicted labels
```

```python
data[1]   #correct labels
```

```python
#comparing graphically
```

```python
sns.scatterplot(feature_1,feature_2, hue = km.labels_) #kmeans wrong clustered in noisy area
```

```python
sns.scatterplot(feature_1,feature_2, hue = data[1])
```

```python
from sklearn import metrics
```

```python
metrics.classification_report(data[1],km.labels_,output_dict=True)
```

```python
metrics.confusion_matrix(data[1],km.labels_)
```
