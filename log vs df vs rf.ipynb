```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('loan_data.csv')
```

```python
df.head()
```

```python
#sns.pairplot(df)
```

```python
plt.figure(figsize=(20,10))
sns.countplot('purpose',data = df, hue = 'not.fully.paid')
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='fico',y = 'int.rate',data = df, hue = 'not.fully.paid')
```

```python
df[['int.rate','fico']].corr()  #indicates good correlation between interest rate and fico
```

```python
df.corr()
```

```python
Purpose = pd.get_dummies(df['purpose'],drop_first=True)
```

```python
Purpose
```

```python
df.drop('purpose',axis=1,inplace=True)
```

```python
df = pd.concat([df,Purpose],axis=1)
```

```python
df.head()
```

```python
#building model
```

```python
X = df.drop('not.fully.paid',axis=1)
y = df['not.fully.paid']
```

```python
#splitting data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#using logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(max_iter=400)
```

```python
lr.fit(X_train,y_train)
```

```python
pred = lr.predict(X_test)
pred
```

```python
y_test
```

```python
#evaluationg performance of model using classification metrics
```

```python
from sklearn import metrics
```

```python
metrics.classification_report(y_test,pred,output_dict=True)
```

```python
metrics.confusion_matrix(y_test,pred)
```

```python
#using decision tree
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
pred2 = dt.predict(X_test)
pred2
```

```python
y_test
```

```python
metrics.classification_report(y_test,pred2,output_dict=True)
```

```python
#using random forest
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=200)
```

```python
rf.fit(X_train,y_train)
```

```python
pred3 = rf.predict(X_test)
pred3
```

```python
y_test
```

```python
metrics.classification_report(y_test,pred3,output_dict=True)
```
