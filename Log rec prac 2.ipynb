```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('titanic_train.csv')
df.head()
```

```python
df.drop('PassengerId',axis = 1, inplace=True
       )
```

```python
df.head()
```

```python
df.drop('Name',axis = 1, inplace=True)
```

```python
df.head()
```

```python
df.drop('Ticket',inplace = True, axis  =1)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop('Cabin',inplace=True, axis = 1)
```

```python
df.head()
```

```python
sns.boxplot(x = 'Pclass', y= 'Age',data = df)
```

```python
def fill(x):
    age = x[0]
    pclass =x[1]
    
    if pd.isnull(age):
        if pclass == 1:
            return 38
        elif pclass == 2:
            return 28
        else:
            return 24
    else:
        return age
```

```python
df['Age'] = df[['Age','Pclass']].apply(fill,axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
sns.countplot(x = 'Survived',data = df, hue = 'Sex')
```

```python
df.hist(column='Fare',by= 'Pclass')
```

```python
Sex = pd.get_dummies(df['Sex'],drop_first=True)
Sex
```

```python
Pclass = pd.get_dummies(df['Pclass'],drop_first=True)
Pclass
```

```python
Embarked = pd.get_dummies(df['Embarked'],drop_first=True)
Embarked
```

```python
df = pd.concat([df,Sex,Embarked,Pclass],axis=1)
df.head()
```

```python
df.drop(['Pclass','Sex','Embarked'],axis = 1 , inplace=True)
```

```python
df.head()
```

```python
X = df.drop('Survived',axis = 1)
y= df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg = LogisticRegression()
```

```python
lg.fit(X_train,y_train)
```

```python
pred = lg.predict(X_test)
```

```python
pred
```

```python
y_test
```

```python
from sklearn import metrics
```

```python
metrics.classification_report(y_test,pred,output_dict=True)
```

```python
metrics.confusion_matrix(y_test,pred)
```
